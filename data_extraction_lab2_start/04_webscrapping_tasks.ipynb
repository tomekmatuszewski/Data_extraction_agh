### Zadanie nr 6

Napisz kod który pobiera i wypisuje na ekran wyniki wszystkich wyścigów Formuły 1 w roku 2018. Punktem wyjścia będzie strona https://www.formula1.com/en/results.html/2018/races.html

Na końcu wyświetl posortowaną listę wszystkich zwycięzców wyścigów wraz z liczbą wygranych wyścigów. Ile kierowców zwyciężało w 2018 roku?

Fragment przykładowego wyniku:
```

   Pos: 1 - Driver: Sebastian Vettel VET
   Pos: 2 - Driver: Lewis Hamilton HAM
   Pos: 3 - Driver: Kimi Räikkönen RAI
   Pos: 4 - Driver: Daniel Ricciardo RIC
   Pos: 5 - Driver: Fernando Alonso ALO
   Pos: 6 - Driver: Max Verstappen VER
   Pos: 7 - Driver: Nico Hulkenberg HUL
   Pos: 8 - Driver: Valtteri Bottas BOT
```

````python
# TO DO zaimplementuj tutaj swoje rozwiązanie
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
from collections import defaultdict

response = requests.get("https://www.formula1.com/en/results.html/2018/races.html").content
soup = BeautifulSoup(response, 'html.parser')
options = [option["value"] for option in  soup.find_all('option')]
options = list(filter(lambda x: re.match(pattern, x),options))
container = defaultdict(list)

for option in options:
    response = requests.get(f'https://www.formula1.com/en/results.html/2018/races/{option}/race-result.html').content
    soup = BeautifulSoup(response, 'html.parser')
    print(f"Race: {option[4:].title()}")
    print(f'Location: {soup.find_all("span", "circuit-info")[0].text}')
    table = soup.find("table", {"class": "resultsarchive-table"})
    table_body = table.find('tbody')
    rows = table_body.find_all('tr')
    
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        name = ' '.join(cols[3].split('\n'))
        container[name].append(cols[1])
        print(f"pos: {cols[1]} - Driver: {name}")
        
        
winners = {k: v.count("1") for k,v in dict(container).items()}
sorted_winners = {k:v for k,v in sorted(winners.items(), key=lambda x: x[1], reverse = True)}

df = pd.DataFrame.from_dict(sorted_winners, orient="index", columns=["win_num"])
df = df[df["win_num"] > 0]
df

````

````python
# TO DO zaimplementuj tutaj swoje rozwiązanie
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
from collections import defaultdict

response = requests.get("https://www.formula1.com/en/results.html/2018/races.html").content
soup = BeautifulSoup(response, 'html.parser')
links = soup.select(".ArchiveLink")
links = [(link["href"], link.text.strip()) for link in links]
container_1 = defaultdict(list)

for link in links:
    url = 'https://www.formula1.com' + link[0]
    print(f"Race: {link[1]}")
    response = requests.get(url).content
    soup = BeautifulSoup(response, 'html.parser')
    print(f'Location: {soup.find_all("span", "circuit-info")[0].text}')
    table = soup.find("table", {"class": "resultsarchive-table"})
    table_body = table.find('tbody')
    rows = table_body.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        name = ' '.join(cols[3].split('\n'))
        container_1[name].append(cols[1])
        print(f"pos: {cols[1]} - Driver: {name}")
               
winners = {k: v.count("1") for k,v in dict(container_1).items()}
sorted_winners = {k:v for k,v in sorted(winners.items(), key=lambda x: x[1], reverse = True)}

df1 = pd.DataFrame.from_dict(sorted_winners, orient="index", columns=["win_num"])
df1 = df1[df1["win_num"] > 0]
df1

````

````python
# import sys
# !"{sys.executable}" -m pip install asyncio --user
# !"{sys.executable}" -m pip install aiohttp --user

import aiohttp
import asyncio

response = requests.get("http://www.sejm.gov.pl/Sejm8.nsf/poslowie.xsp").content
soup = BeautifulSoup(response, 'html.parser')
names = soup.find_all("div", {"class": "deputyName"})
number_of_persons = len(names)

async def get_data(session: aiohttp.ClientSession, id: int):
    if id < 10: id = f"00{id}" 
    elif 10 <= id < 100: id = f"0{id}"    
    url = f"https://www.sejm.gov.pl/Sejm8.nsf/posel.xsp?id={id}&type=P"
    async with session.get(url) as response:
        html = await response.text()
        soup = BeautifulSoup(html, 'html.parser')
        fullname = soup.find_all("h1")[0].text
        date_of_birth = soup.find_all("p", {"id": "urodzony"})[0].text
        num_votes = soup.find_all("p", {"id": "lblGlosy"})[0].next_sibling.text
        group = soup.find_all("p", {"id": "lblLista"})[0].next_sibling.text
        return fullname, date_of_birth, num_votes, group

async def main(number_of_envoys: int):
    async with aiohttp.ClientSession() as session:
        tasks = []
        for id in range(1,number_of_persons+1):
            tasks.append(get_data(session=session, id=id))
        data = await asyncio.gather(*tasks, return_exceptions=True)
        return data
    
result = await main(number_of_envoys)
print(*result, sep="\n")
print(len(result))
        
````
